```python
with open('wikidata.txt') as f:
    lines = f.readlines()
```

```python
wd = []
name = []
file_name = []
foundindices = []
```

```python
for i in range(len(lines)):
    wd.append(lines[i].split('\t')[0])
    name.append(lines[i].split('\t')[1])
    file_name.append(lines[i].split('\t')[2])

for i in range(len(file_name)):
    file_name[i] = file_name[i].replace('commons:','')
    file_name[i] = file_name[i].replace('\n','')
```
