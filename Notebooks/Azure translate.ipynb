```python
import requests, uuid, json

# Add your subscription key and endpoint
subscription_key = ""
endpoint = ""

# Add your location, also known as region. The default is global.
# This is required if using a Cognitive Services resource.
location = ""

path = '/translate'
constructed_url = endpoint + path

params = {
    'api-version': '3.0',
    'from': 'en',
    'to': 'hi'
}

headers = {
    'Ocp-Apim-Subscription-Key': subscription_key,
    'Ocp-Apim-Subscription-Region': location,
    'Content-type': 'application/json',
    'X-ClientTraceId': str(uuid.uuid4())
}
```

```python
path1 = '/transliterate'
constructed_url1 = endpoint + path1

params1 = {
    'api-version': '3.0',
    'language': 'hi',
    'fromScript': 'latn',
    'toScript': 'deva'
}

headers1 = {
    'Ocp-Apim-Subscription-Key': subscription_key,
    'Ocp-Apim-Subscription-Region': location,
    'Content-type': 'application/json',
    'X-ClientTraceId': str(uuid.uuid4())
}
```

```python
def my_translate(val):
    #  if (val == ''):
    #          t1 = ''
    #  else:  
    #          a = anu.anuvaad(val)
    #          t1 = transliterate_numerals(a, lang_code='hi')
            
    # e = model.translate(val, target_lang='hi')
    # t1 = transliterate(e, sanscript.IAST, sanscript.DEVANAGARI)
    
    # You can pass more than one object in body.
    body = [{'text': val}]
    request = requests.post(constructed_url, params=params, headers=headers, json=body)
    response = request.json()
    t1 = response[0]['translations'][0]['text']
    return t1

def my_transliterate(val):
    # You can pass more than one object in body.
    body = [{'text': val}]
    request = requests.post(constructed_url1, params=params1, headers=headers1, json=body)
    response = request.json()
    t2 = response[0]['text']
    return t2
```

```python
vect_translate = np.vectorize(my_translate)
vect_transliterate = np.vectorize(my_transliterate)
```
